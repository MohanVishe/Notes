# Python Functions

A function in Python is a group of related statements that perform a specific task. Functions help break a program into smaller and more manageable chunks, making the code organized, reusable, and preventing repetition.

#### Syntax
```python
def function_name(parameters):
    """Doc String"""
    Statement(s)
```

1. `def`: Keyword marking the start of the function header.
2. `parameters`: Optional, through which we pass values to a function.
3. `:`: Marks the end of the function header.
4. `Doc String`: Describes what the function does (optional).
5. `return` statement: Optional, used to return a value from the function.


#### Example

````python
def print_name(name):
    """This function prints the name"""
    print("Hello " + str(name))
````

#### Function Call
Once defined, we can call a function from anywhere in the code:

````python
print_name('satish')
````

#### Return Statement
The `return` statement is used to exit a function and return a value to the calling code. It can contain an expression that gets evaluated and returned. If no expression is present, or the `return` statement is absent, the function returns `None`.

````python
def get_sum(lst):
    """This function returns the sum of all elements in a list"""
    _sum = 0
    for num in lst:
        _sum += num
    return _sum
````

````python
get_sum([1, 2, 3, 4])
````

#### Example: Computing Highest Common Factor (HCF)

````python
def computeHCF(a, b):
    """Computing HCF of two numbers"""
    smaller = b if a > b else a
    hcf = 1
    for i in range(1, smaller + 1):
        if (a % i == 0) and (b % i == 0):
            hcf = i
    return hcf

print("H.C.F of {0} and {1} is: {2}".format(6,36, computeHCF(6,36)))
````

## Functional Arguments

1. **Positional Arguments:**

   Positional arguments are the most common type of function arguments. They are defined in the function signature and must be provided in the same order when calling the function.

````python
def add(a, b):
  return a + b
````

````python
add(2, 3)
# Here, 2 and 3 are positional arguments
````

2. **Keyword Arguments:**

   Keyword arguments are specified by name when calling a function. They allow you to pass arguments out of order, making the function call more readable.

````python
def greet(name, message):
  return f"Hello, {name}! {message}"
````

````python
greeting = greet(message="Welcome to Python", name="Alice")
````

3. **Default Arguments:**

   Default arguments have predefined values in the function signature. If a value is not provided for a default argument during the function call, it uses the default value.

````python
def greet(name, message="Welcome"):
  return f"{message}, {name}!"
````

````python
greet("Bob")  # Uses default message: "Welcome, Bob!"
````

````python
greet("Alice", "Hello")  # Overrides the default message
````

4. **Arbitary Arguments:**

   Sometimes, we do not know in advance the number of arguments that will be passed into a function.Python allows us to handle this kind of situation through function calls with arbitrary number of arguments.

  - **Arbitrary Positional Arguments:**

   You can use the `*args` syntax to accept an arbitrary number of positional arguments. These arguments are collected into a tuple.

````python
def sum_all(*args):
  result = 0
  for num in args:
    result += num
  return result
````

````python
sum_all(20,5,10)
````

````python
sum_all(200,100,10,20,20)
````


  - **Arbitrary Keyword Arguments:**

   You can use the `**kwargs` syntax to accept an arbitrary number of keyword arguments. These arguments are collected into a dictionary.


````python
def print_info(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")
````

````python
print_info(name="Alice", age=30, city="New York")
# Accepts any keyword arguments
````

# Recursive Function


- A function can call other functions. It is even possible for the function to call itself. These type of construct are termed as recursive functions.


Advantages of Recursive Functions:
- Recursive functions can provide elegant and clean solutions to problems that involve repetitive or recursive structures.
- They can break down complex tasks into simpler sub-problems, making the code more manageable.

Disadvantages of Recursive Functions:
- Understanding and following the logic behind recursion can be challenging for complex cases.
- Recursive calls can be less efficient in terms of memory and time complexity compared to iterative solutions.
- Debugging recursive functions can be difficult, especially with deep recursion.


**Example:** Factorial

````python
def factorial(num):
  """Recursive function to find factorial"""
  if num==1:
    return 1
  else :
    return num * factorial (num-1)
````

````python
factorial(5)
````

### **Steps create recursive function**

1. Base condition , Termination condition

2. Logic

3. Recurssion condition

#### **Example 1**: print given list

  * Normal way




````python
lst=[1,2,3,4]

for i in lst:
  print(i)
````

  * recurssion

````python
def looping(lst,i=0):
  if i<len(lst):                   # Base/Termination condition
    print(lst[i])                  # Logic
    looping(lst,i+1)               # Recurssion condition
````

#### **Example 2** :  print given list in reverse order

  * Normal way

````python
lst=[1,2,3,4]

i=len(lst)-1
for j in range(len(lst)):
  print(lst[i])
  i=i-1
````

* recurssion

````python
# we can change the order of recurssion function and print in reverse

def revloop(lst,i=0):
  if len(lst)>i:              # Termination condition
    revloop(lst,i+1)          # recurssion condition
    print(lst[i])             # Logic
````

### **Stack**



image.png

- Explanation of how recursion uses the call stack data structure.

1. **Function Call and Stack Frame Creation**:
   - When a function is initially called, a new stack frame is created on the call stack to store information about that function call.
   - This stack frame contains data like the function's local variables, parameters, and the memory address of the instruction in the code where the function call was made.

2. **Recursive Calls**:
   - Within the function's code, if there is a recursive call to the same function, a new stack frame is created for the new call, and it's pushed onto the top of the call stack.
   - Each recursive call has its own set of local variables and parameters, distinct from the previous calls. This allows each call to work with its own data independently.

3. **Storing State**:
   - Each stack frame on the call stack stores the context of a specific function call. This context includes:
     - Local variables and their values.
     - Parameters and their values.
     - The return address, which is the location in the code where execution should continue once the current call is complete.

4. **Recursive Call Execution**:
   - The new stack frame begins executing the recursive call with its specific set of local variables and parameters.
   - If the recursive call itself contains further recursive calls, this process continues, creating more stack frames.

5. **Returning from Recursive Calls**:
   - As the base case is reached, or a condition is met that signals the termination of the recursion, the current stack frame starts to return.
   - The return address in the stack frame is used to resume execution in the caller function, and the local variables of that frame are popped off the stack.

6. **Popping Call Stack Frames**:
   - The call stack follows a Last-In-First-Out (LIFO) order. As each function call completes, its stack frame is removed from the top of the call stack.
   - This process continues until all the stack frames are removed, signifying the end of the recursive process.

7. **Resource Limitations**:
   - It's important to be aware that the call stack is a finite resource. If recursion goes too deep or consumes too much stack space, it can lead to a stack overflow error. Therefore, it's crucial to design recursive functions with a well-defined base case to ensure that the recursion eventually stops and to manage stack space efficiently.

- Recursion uses the call stack to manage the order of function calls, allowing each call to maintain its own set of data. This mechanism allows for the execution of recursive functions, where a function calls itself, until a base case is met, at which point the calls return and the call stack frames are popped, one by one. Care must be taken to prevent stack overflow errors by properly designing recursive functions and handling the termination condition effectively.


# Generator Functions:



1. **Definition and Purpose:**
    * A generator function is defined like a regular function but uses the `yield` statement instead of `return`. When called, it doesn't execute the function immediately but returns a generator object.
    * The primary purpose is to generate a sequence of values on-the-fly, saving memory as values are generated one at a time.

2. **Iterator:**
    * A generator function returns an iterator.
    * An iterator is an object with two important methods, `__iter__()` and `__next__()`, which allow you to traverse through a sequence of values.

### Example 1: Simple Generator Function



````python
def simple_generator():
    yield "mohan"
    yield "gaurav"
    yield "rahul"

gen = simple_generator()
````

````python
print(next(gen))
````

````python
print(next(gen))
````

````python
print(next(gen))
````

In this example, `simple_generator` generates values 1, 2, and 3 one at a time using `yield`.

## Benefits of Generator Functions:

3. **Memory Efficiency:** Generator functions produce values lazily, meaning they generate and yield each value only when requested. This is especially useful for large datasets or infinite sequences where storing all values in memory would be impractical.

### Example 2: Generating an Infinite Sequence

````python
def infinite_sequence():
    num = 1
    while True:
        yield num
        num += 1

gen = infinite_sequence()
print(next(gen))
print(next(gen))
# You can keep calling next(gen) indefinitely to generate more numbers.
````

4. **Efficiency:** Generator functions can be more efficient than storing all values in memory, especially when dealing with calculations or data processing where each value is derived from the previous one.

### Example 3: Fibonacci Sequence

````python
def fibonacci_sequence():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

gen = fibonacci_sequence()
for _ in range(10):
    print(next(gen))
````

5. **Easily Composable:** Generator functions can be composed together using iterators, allowing you to create complex data processing pipelines with minimal memory overhead.

### Example 4: Composing Generators

````python
def squares(n):
    for i in range(n):
        yield i ** 2
````

````python
def even_squares(n):
    for num in squares(n):
        if num % 2 == 0:
            yield num

gen = even_squares(10)
for val in gen:
    print(val)
#(Squares of even numbers up to 10)
````
