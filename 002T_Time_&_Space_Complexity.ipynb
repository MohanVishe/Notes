# Time and space complexity

* Time and space complexity are two fundamental concepts in computer science that help analyze and compare the efficiency of algorithms.
* They provide a way to measure how an algorithm's performance scales with respect to **input size**.

## **Time Complexity:**
-  Time complexity is a measure of the amount of **time an algorithm** takes to run as a function of the **size of its input** for constant programming language, computation power, OS,etc  
- It helps us understand how the running time of an algorithm grows with increasing input size.
- Time complexity is typically expressed using Big **` O `** notation, which describes the upper bound or worst-case scenario of an algorithm's running time.

Common time complexities:

1. **O(1) - Constant Time**:
- An algorithm takes a constant amount of time to complete, regardless of the input size.
- It is the most efficient time complexity.

- Example 1

```python
def constant_time_example(arr):
    return arr[0]
```

```python
# Regardless of the input size, this function takes constant time.
O(1)
```

2. **O(n) - Linear Time**: The running time of the algorithm increases linearly with the input size. For each additional input element, the algorithm performs a constant amount of work.

```python
def linear_time_example(arr):
    total = 0
    for num in arr:
        total += num             # "O(n)"
    return total

# The time taken is proportional to the length of 'arr'.
```

3. **O(n^2) - Quadratic Time**:
- The running time of the algorithm is proportional to the square of the input size.
- Commonly seen in nested loops.

- Example 1

```python
def quadratic_time_example(arr):
    n = len(arr)
    result = 0
    for i in range(n):
        for j in range(n):
            result += arr[i] * arr[j]   # "O(n) * O(n)"
    return result
```

```python
# The time taken grows quadratically with the length of 'arr'.
O(n**2)
```

- Example 2

```python
for i in range(n):
  print(i)               # "O(n)"
  for j in range(n):
    print(j)             # "O(n) * O(n)"= O(n**2)
```

```python
# taking maximum time complexity into account
# (while adding we take maximum time complexity)
O(n)+O(n**2)
=O(n**2)
```

## **Space Complexity:**
- Space complexity is nothing but the amount of memory space that an algorithm or a problem takes during the execution of that particular problem/algo.

- The space complexity is not only calculated by the space used by the variables in the problem/algo it also includes and considers the space for input values with it.

- There is a sort of confusion among people between the space complexity and the auxiliary space. So let’s be clear about that, so auxiliary space is nothing but the space required by an algorithm/problem during the execution of that algorithm/problem and it is not equal to the space complexity because space complexity includes space for input values along with it also.

- So we can say that space complexity is the combination or sum up of the auxiliary space and the space used by input values.

- **Space Complexity** = **Auxiliary Space** + **Space used for input values**


* Example

```python
def constant_space_example(n):
    a = 5
    b = 10
    return a + b

# This function uses only a fixed amount of memory.
```

```python
# n is "int" takes space O(1) [space used for input values]
# a,b are "int" takes memory O(1) [Auxiliary Space]
# so space complexity is O(1)+O(1)+O(1)=O(1)
```

* Example

```python
# Swapping two variables using a temporary variable
def swap_variables(a, b):
    temp = a
    a = b
    b = temp

x = 5
y = 10
swap_variables(x, y)
```

```python
# space take by a,b,temp is O(1)
# space complexity is O(1)
```

```python
# Storing an input array in another array
def duplicate_array(arr):   # arr=O(n)
    new_arr = []
    for item in arr:        # item= O(1)
        new_arr.append(item)

    return new_arr          # new_arr=O(n)

input_list = [1, 2, 3, 4, 5]
duplicated = duplicate_array(input_list)
```

```python
# space complexity = O(n)+O(1)+O(n) = O(n)
```

* Example

```python
#  Performing a depth-first search (DFS) on a tree
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.children = []

def dfs(node):
    visited = set()
    stack = [node]

    while stack:
        current = stack.pop()
        if current not in visited:
            visited.add(current)
            stack.extend(current.children)
```

```python
# Implementing a queue using an array
class Queue:
    def __init__(self):
        self.items = []    # self.item = O(n)

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)

    def is_empty(self):
        return len(self.items) == 0
```

## General overview of the worst-case, average-case, best-case time and space complexity for common time complexity classes in algorithm analysis:

**Time Complexity:**

1. **Best-Case Time Complexity (O(Best)):** This represents the minimum amount of time an algorithm takes to complete when given the best possible input. It reflects the most efficient performance of the algorithm.

2. **Average-Case Time Complexity (O(Avg)):** This represents the average time an algorithm takes to complete when given an input randomly chosen from all possible inputs. It considers all possible input scenarios and provides an average estimate of an algorithm's performance.

3. **Worst-Case Time Complexity (O(Worst)):** This represents the maximum amount of time an algorithm takes to complete when given the worst possible input. It describes the scenario in which the algorithm performs least efficiently.

**Space Complexity:**

1. **Best-Case Space Complexity (O(BestSpace)):** This represents the minimum amount of memory an algorithm uses when given the best possible input. It reflects the most efficient use of memory by the algorithm.

2. **Average-Case Space Complexity (O(AvgSpace)):** This represents the average amount of memory an algorithm uses when given an input randomly chosen from all possible inputs. It considers all possible input scenarios and provides an average estimate of an algorithm's memory usage.

3. **Worst-Case Space Complexity (O(WorstSpace)):** This represents the maximum amount of memory an algorithm uses when given the worst possible input. It describes the scenario in which the algorithm uses the most memory.

Suppose you have a sorted list and an integer, and you need to determine if the integer exists in the list. A straightforward approach to solving this problem is to traverse the entire list and check if any element is equal to the given integer.


```python
integer_to_match = 13
list_to_search = [1, 3, 5, 8, 12, 13, 15, 16, 18, 20, 22, 30, 40, 50, 55, 67]

for i in range(len(list_to_search)):
    if integer_to_match == list_to_search[i]:
        print("Match found at index:", i)
```

This linear search has a time complexity of O(n).

However, a more efficient solution can be achieved using binary search, which has a time complexity of O(log n):

```python
def binary_search(arr, low, high, x):
    if high >= low:
        mid = (high + low) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            return binary_search(arr, low, mid - 1, x)
        else:
            return binary_search(arr, mid + 1, high, x)
    else:
        return -1

integer_to_match = 13
list_to_search = [1, 3, 5, 8, 12, 13, 15, 16, 18, 20, 22, 30, 40, 50, 55, 67]

result = binary_search(list_to_search, 0, len(list_to_search) - 1, integer_to_match)
if result != -1:
    print("Match found at index", result)
```

*  Find elements common in two lists:

```python
import random
l1 = list(range(100))
random.shuffle(l1)
l2 = list(range(50))
random.shuffle(l2)
# find common elements : time complexity O(n*m)
cnt=0;
for i in l1:
  for j in l2:
    if i==j:
      cnt += 1;
print("Number of common elements:", cnt)
```

* we reduce time complexity by using dictionary

```python
# find common elemnts in lists in O(n) time and O(m) space if m<n
## add all elements in the smallest list into a hashtable/Dict: O(m) space
smallList = {}
for ele in l2:
  smallList[ele] = 1; # any value is OK. Key is important

# Now find common element
cnt=0;
for i in l1:
  if smallList.get(i) != None:
    # search happens in constant time.

    cnt += 1;
print("Number of common elements:", cnt)
```

```python
# as we used the "dict" if find instant O(1) so O(1)*O(m)
# time complexity become O(n)+O(m) as n>m the time complexity is O(n)
# because dictionary detect O(1) time for finding element but it took space O(m)
```
