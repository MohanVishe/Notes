```python
# Python lists
L = range(1000)
%timeit [i**2 for i in L]

# NumPy arrays
a = np.arange(1000)
%timeit a**2
```

### Slicing

- Slicing allows selecting a portion of the array.

```python
a = np.arange(10)
print(a)
```

```python
print(a[1:8:2])  # [start:index(exclusive):step]
```

```python
a[5:] = 10
print(a)
```

- **NOTE (Copies and Views):**  Slicing creates a view on the original array, which is a way of accessing array data without copying it


```python
a = np.arange(10)
b = a[::2]  # This creates a view of the original array

# Check if a and b share the same memory
print(np.shares_memory(a, b))
```

```python
# Use .copy() to avoid this problem
c= a[::2].copy()
print(np.shares_memory(a,c))
```

## **Broadcasting in NumPy**

**Basic Operations on Numpy Arrays**
- Basic operations in NumPy, such as addition, subtraction, multiplication, and division, are performed elementwise.
- These operations work on arrays of the same size, but NumPy also allows broadcasting to perform operations on arrays of different sizes.

**Broadcasting in NumPy**
- Broadcasting is a way to perform operations on arrays of different sizes by transforming them to have the same size.
- Broadcasting follows a set of rules:
    1. If the arrays have different dimensions, pad the smaller dimension with ones on the left.
    2. If the shape of the arrays doesn't match in a dimension, the array with shape equal to 1 is stretched.
    3. If sizes disagree and neither is equal to 1, an error is raised.


### **Examples:**

1. Broadcasting with Scalars:

```python
np.arange(3)
```

```python
np.arange(3) + 5
```

2. Broadcasting with Arrays:

```python
np.ones((3, 3))
```

```python
np.ones((3, 3)) + np.arange(3)
```

3. Broadcasting with Different Shaped Arrays:

```python
np.arange(3).reshape((3, 1))
```

```python
np.arange(3).reshape((3, 1)) + np.arange(3)
```

4. Broadcasting with Mismatched Dimensions:
   - Broadcasting a 2D array with a 1D array requires care, as dimension matching is crucial.
   - Example: Broadcasting a `(2, 4)` array with a `(2, 2)` array raises an error.

5. Element-Wise Multiplication:
   - `np.multiply(a, b)` multiplies corresponding elements of arrays `a` and `b` element-wise.

6. Element-Wise Division:
   - `np.divide(a, b)` divides corresponding elements of arrays `a` by elements of `b`, which can result in division by zero.


## **Array Methods and functions**
- Operations for reshaping and resizing arrays are essential in NumPy.

1. **Flattening**
  - `a.ravel()` returns a contiguous flattened array.

2. **Reshaping**
  - `b = a.reshape((2, 3))` reshapes the array `a` into a 2x3 array.
  - Beware that `reshape` may return a copy if not used carefully.

3. **Adding a Dimension**
  - Indexing with `np.newaxis` adds an axis to an array, increasing its dimension.

4. **Dimension Shuffling**
  - Arrays can be reshaped to different dimensions without changing the data.
  - Example: `a = np.arange(4*3*2).reshape(4, 3, 2)` reshapes a 3D array.

5. **Resizing**
  - `a.resize((8,))` resizes the array to the specified shape.
  - Beware of referencing or copying arrays when using `resize`.

6. **Sorting Data**
  - NumPy provides methods for sorting along axes or in-place sorting.
  - Sorting with fancy indexing allows you to retrieve sorted indices.

6. **Additional Operations**
  - `np.sum()`, `np.min()`, and `np.max()` for aggregation.
  - Sorting with `np.sort()`.

7. **Array Manipulation Functions**
  - NumPy offers functions like `np.transpose()`, `np.reshape()`, `np.resize()`, and `np.flatten()` for array manipulation.

8. **Manipulating Arrays**
  - `np.insert(arr, obj, values, axis)` allows insertion of values along specified axes.
  - `np.append(arr, values, axis)` appends values along specified axes.
  - `np.delete(arr, obj, axis)` deletes elements specified by `obj` along the specified axis.
  - `np.unique(arr, return_index, return_inverse, return_counts)` returns unique elements from an array with optional additional information.
