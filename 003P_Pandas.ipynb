- Apply Method
  - `df.apply(function,axis)`
  - Create new columns (axis=1) by using function on old columns values row wise

- eval method and function:
  - `df.eval("condition")`  :  to filter rows based on a condition , (condition will be in str format)
  - **`eval()`**  function : eval() is a powerful, yet potentially risky, built-in function in Python. It allows you to evaluate a string as a Python expression or code. While it can be handy for dynamically executing code, it's essential to be cautious because it can introduce security risks if used with untrusted input.

```python
# Create new column
df["birth"]= 1900 +df["year"]-df["age"]
df.tail()
```

```python
# Use Apply method

def age_in_months(y):
  return 12 * y

df["age_M"]=df["year"].apply(age_in_months)
df.head()
```

```python
# Apply with lambda function

df["birth"]=df.apply(lambda x: 1900+x["year"]-x["age"] , axis=1)
df.head()
```

### **2. pandas Library:**

-  Converts a column to datetime format.
  - `pd.to_datetime()`

- Extracting components:

   
  - `df['year'] = df['date'].dt.year`
  - `df['month'] = df['date'].dt.month`
  

- Generates a range of dates.:
  - `pd.date_range(start,period,freq)`
  






```python
df=pd.DataFrame([("02/02/21","Mon",60),("20/12/12","Tue",70),("06/06/06","Wed",65)],
                columns=["date","day","percentage"],index=[1,2,3])
df.info()
```

```python
# convert column datatype to datetime format
df["date"]=pd.to_datetime(df["date"])
df.info()
```

```python
# Extract day,month,year
df["year"]=df["date"].dt.year
df["month"]=df["date"].dt.month
df["day"]=df["date"].dt.day
df
```

### **3. `strftime` and `strptime`**:

- `strftime`: Converts datetime to a string with a specified format.

```python
current_datetime=datetime.datetime.now()
formatted_date = current_datetime.strftime('This is the Date : %Y-%m-%d  & Time : %H:%M:%S')
print(formatted_date)
```

- `strptime`: Parses a string into a datetime object using a specified format.

```python
parsed_datetime = datetime.datetime.strptime('2023-11-09', '%Y-%m-%d')
print(parsed_datetime)
# for above example
parsed_datetime = datetime.datetime.strptime(formatted_date, 'This is the Date : %Y-%m-%d  & Time : %H:%M:%S')
print(parsed_datetime)
```



### 4. `timedelta`:

- `datetime.timedelta(days=x, hours=y, minutes=z)`: Represents a duration of time.


```python
time_difference = datetime.timedelta(days=5, hours=3)
new_datetime = current_datetime + time_difference
print(new_datetime)
```
