# **Matplotlib**


## **Use Matplotlib for data analysis purposes:**

1. **Import Matplotlib**:
   To get started, you need to import Matplotlib. Typically, you would import the `pyplot` module, which provides an interface similar to MATLAB for creating plots and charts.

````python
#install package
!pip install matplotlib
# import
import matplotlib.pyplot as plt
````

2. **Create Data**:
   Before creating visualizations, you'll need some data to analyze. You can use NumPy or other data manipulation libraries to create or import your data.

3. **Basic Plotting**:
   Matplotlib provides various types of plots, including line plots, scatter plots, bar plots, and more. Here's how to create a simple line plot:

````python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]
plt.plot(x, y)
plt.show()
````

4. **Customizing Plots**:
   You can customize your plots using various functions. Some common customizations include:
   - Adding labels to axes:
     ```python
     plt.xlabel('X-axis Label')
     plt.ylabel('Y-axis Label')
     ```
   - Setting a title:
     ```python
     plt.title('My Plot Title')
     ```
   - Adding legends:
     ```python
     plt.legend(['Data Series 1'])
     ```

5. **Subplots**:
   You can create multiple plots within a single figure using subplots. This is useful for comparing different datasets or visualizing data side by side.

````python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'b*')
plt.show()
````

6. **Saving Figures**:
   You can save your plots as image files for use in reports or presentations. Use `savefig` to do this:
   ```python
   plt.savefig('my_plot.png')
   ```

7. **Advanced Plot Types**:
   Matplotlib supports a wide range of plot types, including histograms, bar charts, pie charts, and 3D plots. Explore the Matplotlib documentation for more options.

8. **Data Visualization Best Practices**:
   - Choose the appropriate plot type for your data.
   - Label axes and provide meaningful titles.
   - Use colors and markers to distinguish data points.
   - Ensure your plots are easy to interpret and aesthetically pleasing.

9. **Integration with Jupyter Notebooks**:
   If you're using Jupyter Notebooks, you can display Matplotlib plots inline by using `%matplotlib inline` magic command.

10. **Resources**:
    Matplotlib has extensive documentation and numerous tutorials online. Explore these resources to enhance your data visualization skills.


##  **Creating Basic Plots**

**1. Line Plot:**

````python
plt.plot(x,y,'ro',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('2d diagram')
plt.show()
````

**2. Scatter Plot**


````python
plt.scatter(x,y,c="r")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatter Plot")
plt.show()
````

**3. Bar Chart**



````python
x1=[2,3,4,5]
y1=[11,16,19,20]

x2=[3,9,11,14]
y2=[6,15,7,9]


plt.bar(x1,y1,color='g')
plt.bar(x2,y2)
plt.show()
````

**4. Histogram**


````python
import numpy as np
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=5)
plt.show()
````

**5. Boxplot**

````python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data)
plt.show()
````

**6. Pie Chart**

````python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
````

# **Seaborn**

### **Introduction to Seaborn**
- Seaborn is a powerful Python data visualization library built on top of Matplotlib.
- It provides a high-level interface for creating informative and attractive statistical graphics.

**1. Importing Seaborn**
- To use Seaborn, import it alongside Matplotlib:

````python
#installation
!pip install seaborn
# import
import seaborn as sns
import matplotlib.pyplot as plt
````


**2. Loading Data**
- Seaborn works seamlessly with Pandas DataFrames.
- Use `sns.load_dataset()` or `pd.read_csv()` to load your data.

**3. Setting Aesthetic Parameters**
- Customize the appearance of your plots using Seaborn's aesthetic parameters.
- Use `sns.set_style()` to set a style: 'whitegrid', 'darkgrid', 'ticks', etc.
- Adjust the context with `sns.set_context()`: 'notebook', 'paper', 'talk', or 'poster'.
- Set color palettes with `sns.set_palette()`.
  
**4. Creating Basic Plots**
- Seaborn makes it easy to create a variety of basic plots:
  - Scatter plots: `sns.scatterplot()`
  - Line plots: `sns.lineplot()`
  - Bar plots: `sns.barplot()`
  - Histograms: `sns.histplot()`
  - Box plots: `sns.boxplot()`
  - Violin plots: `sns.violinplot()`

**5. Customizing Plots**
- Customize your plots using Seaborn's functions and arguments.
- Add titles and labels using `plt.title()`, `plt.xlabel()`, and `plt.ylabel()`.
- Modify colors, styles, and markers as needed.
- Use `plt.legend()` to add legends to your plots.

**6. Advanced Visualization**
- Seaborn excels at advanced visualizations:
  - Heatmaps: `sns.heatmap()`
  - Pair plots: `sns.pairplot()`
  - Facet grids: `sns.FacetGrid()`
  - Joint plots: `sns.jointplot()`
  
**7.  Working with Categorical Data**
- Seaborn offers tools for categorical data visualization:
  - Count plots: `sns.countplot()`
  - Categorical scatter plots: `sns.stripplot()`, `sns.swarmplot()`
  - Categorical box plots: `sns.boxplot()`
  
**8. Data Exploration**
- Use Seaborn's functions like `sns.distplot()`, `sns.pairplot()`, and `sns.heatmap()` to explore data distributions and relationships.

**9. Saving Plots**
- Save your Seaborn plots using `plt.savefig("filename.png")`.
