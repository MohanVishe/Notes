## **Expressions and Statements in Python:**

**Expression:**
An expression in Python is a combination of values, operators, and/or functions that can be evaluated to produce a result. Expressions can be simple or complex and can involve various types of data.

- **Components of an Expression:**
  - **Values:** These are the data elements used in an expression. They can be constants like strings ("hello"), numbers (-87.8, 6), or variables.
  - **Operators:** These are symbols that perform operations on values. Common operators include `*` (multiplication), `-` (subtraction), `/` (division), and `+` (addition).
  - **Functions:** These are built-in or user-defined operations that can be applied to values to produce a result.

- **Purpose of Expressions:**
  Expressions are used to perform calculations, comparisons, or other operations to generate new values. They are a fundamental building block of Python programs and are often used in assignments, conditional statements, and loops.

**Example Expression:**
```python
result = (19 - 2) * 6
```
In this example:
- The expression `(19 - 2)` calculates the value `17`.
- The expression `(19 - 2) * 6` calculates the value `102`.
- The operators `-` and `*` perform subtraction and multiplication operations on the values.

**Statement:**
A statement in Python is a complete line of code that performs a specific action or task. It can be a single expression or a group of expressions that accomplish a particular purpose.

- **Examples of Statements:**
  - Assignment Statement: `spam = 10`
  - Print Statement: `print("Hello, World!")`
  - Conditional Statement: `if x > 5:`
  - Looping Statement: `for i in range(5):`

- **Expression as Part of a Statement:**
  Expressions often play a role in statements. For example, in the assignment statement `spam = 10`, the expression `10` is assigned to the variable `spam`.

## Data types and Data Structures

**Data Types:**

Data types in Python define the kind of values that can be stored and manipulated in a program. Python provides several built-in data types:

6. **Sequence Types:**

7. **Set Types:**
   - **set:** Represents an unordered collection of unique items, e.g., `{1, 2, 3}`.

## Flow Control in Python

### if-else Statements

The `if-else` statement is used to execute different code blocks based on specific conditions. The syntax is as follows:

````python
if condition1:
    # Code block to execute if condition1 is True
elif condition2:
    # Code block to execute if condition2 is True
else:
    # Code block to execute if none of the conditions are True
````

image.png

- The test expressions or conditions are evaluated, and the corresponding code block is executed if the condition is true.
- If none of the conditions are true, the code block under the `else` statement is executed.

*  **Example of if-else Statement:**

````python
# Checking whether a number is positive, negative, or zero
num = int(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
````

In this example, the user enters a number, and the code uses `if-else` statements to check whether the number is positive, negative, or zero.

* **Example of Nested if-else Statement:**


````python
# Classifying a student's performance based on marks
marks = int(input("Enter the student's marks: "))

if marks >= 90:
    grade = 'A'
    if marks >= 95:
        remark = "Excellent"
    else:
        remark = "Very Good"
elif marks >= 80:
    grade = 'B'
    remark = "Good"
elif marks >= 70:
    grade = 'C'
    remark = "Average"
elif marks >= 60:
    grade = 'D'
    remark = "Below Average"
else:
    grade = 'F'
    remark = "Fail"

print(f"Grade: {grade}")
print(f"Remark: {remark}")
````

### `break` Statement

The `break` statement is used to exit a loop prematurely. When the interpreter encounters `break`, it immediately exits the loop.






image.png

````python
for i in range() :
else :
# when for loop will complete else will executed if there is break then not
````

* **Example 1: Check if a Number is Prime or Not (Without `break`)**

````python
number = int(input("Enter a number:"))

if number == 1:
    print("1 is neither prime nor composite.")
elif number == 2:
    print("2 is a prime number.")
else:
    i = 2
    while i < number:
        if number % i == 0:
            print("{} is divisible by {} hence it is not a prime number.".format(number, i))
        i += 1
    else:
        print("{} is a prime number.".format(number))
````

* **Example 2: Check if a Number is Prime or Not (With `break`)**





````python
number = int(input("Enter a number:"))

if number == 1:
    print("1 is neither prime nor composite.")
elif number == 2:
    print("2 is a prime number.")
else:
    i = 2
    while i < number:
        if number % i == 0:
            print("{} is divisible by {} hence it is not a prime number.".format(number, i))
            break
        i += 1
    else:
        print("{} is a prime number.".format(number))
````

In this example, the code also checks if the given number is prime or not using a `while` loop. However, it includes a `break` statement within the loop. This means that if the number is found to be divisible by any value within the range, the loop immediately terminates, and the number is declared as not prime. If the loop completes without encountering the `break` statement, the number is prime.

• False
