# **Mongo DB**

## **Information About Mongo DB**





-  MongoDB is a NoSQL, document-oriented database, and it doesn't enforce a rigid schema like SQL databases do.
- Instead, MongoDB uses a flexible and dynamic schema, allowing documents within a collection to have different fields.



1. **Schema-less Design:**

   - MongoDB is schema-less, meaning each document in a collection can have a different structure.
   - Documents in the same collection are not required to have the same fields, and fields can vary in type.

2. **Fields:**

   - Fields in MongoDB are similar to columns in a relational database.
   - Each document consists of key-value pairs known as fields.

3. **Data Types:**

   - MongoDB supports various data types for fields, including string, number, boolean, array, embedded document, and more.
   - Data types can be nested, allowing for the creation of complex and rich data structures.

4. **Dynamic Schema:**

   - Fields can be added to documents on the fly.
   - There's no need to predefine the structure of documents or the schema of a collection.

5. **Default Values:**

   - MongoDB allows specifying default values for fields, which are used if a field is not present in a document.

 6. **Validation Rules:**

   - Starting from MongoDB version 3.2, you can enforce validation rules on fields.
   - Validation rules ensure that documents meet specific criteria before they are inserted or updated in a collection.

7. **Subdocuments and Arrays:**

   - Fields in MongoDB can contain subdocuments or arrays, enabling the representation of hierarchical data structures.
   - This flexibility allows the storage of complex and nested information within a single document.

 8. **Indexes:**

   - While MongoDB is schema-less, it supports the creation of indexes on fields to improve query performance.
   - Indexes can be created on single fields, compound fields, or even on arrays.

9. **Schema Design Considerations:**

   - Schema design in MongoDB depends on the specific requirements of your application.
   - Considerations include the types of queries your application will perform and the structure of the data.

- Example:

MongoDB document representing a person:

```json
{
  "_id": ObjectId("5f98a48c3b07a29f3826b7a1"),
  "name": "Alice",
  "age": 30,
  "email": "alice@example.com",
  "address": {
    "city": "Wonderland",
    "country": "Fictionland"
  },
  "hobbies": ["reading", "gardening"]
}
```

Document has various fields like `name`, `age`, `email`, `address`, and `hobbies`. The `address` field is a subdocument, and the `hobbies` field is an array.

### **Inserting Record into Collection**

````python
# Creating record
firstrecord={"name":"mohan","lname":"vishe","address":"mumbai"}

# Insert record into collection
mycol.insert_one(firstrecord)
````

````python
# Creating record
secondrecod={"name":"gaurav","lname":"yadav","address":"new delhi","Salary":4000}

# Insert record into collection
mycol.insert_one(secondrecod)
````

````python
# Creating multiple records

multiplerecord=[{"name":"rahul","lname":"ray","address":"new delhi"},
                {"name":"shambhuraj","lname":"desai","address":"satara"},
                {"name":"raj","lname":"singh","address":"pune"}]
mycol.insert_many(multiplerecord)
````

### **Updating Documents/Records**
    - update_one()  : only update one document even if there are multiple document with same query
    - update_many() : update all documents which satisfy the given query

````python
mycol.update_one({"name":"gaurav","lname":"yadav","address":"new delhi","Salary":4000},
                 {"$set": {"address":"mumbai","Salary":90000}})
````

````python
mycol.update_many({"name":"gaurav"},
                  {"$set":{"address":"USA"}})
````

### **Deleting Documents/Records**

````python
mycol.delete_one({"name":"rahul"})
````

````python
mycol.delete_many({"name":"raj"})
````

### **Querying Documents/Records**



| Querying Aspect                                 | Explanation                                                                                                                                                                                           | Example                                                   |
|--------------------------------------------------|-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|-----------------------------------------------------------|
| Basic Find                                       | Retrieve documents from a collection.                                                                                                                                                                | `collection.find()`                                       |
| Equality                                         | Retrieve documents where a specified field is equal to a certain value.                                                                                                                            | `collection.find({"field": value})`                       |
| Inequality                                       | Retrieve documents where a specified field is not equal to a certain value.                                                                                                                        | `collection.find({"field": {"$ne": value}})`            |
| Logical AND                                      | Combine multiple conditions using the logical AND operator.                                                                                                                                         | `collection.find({"field1": value1, "field2": value2})` |
| Logical OR                                       | Combine multiple conditions using the logical OR operator.                                                                                                                                          | `collection.find({"$or": [{"field1": value1}, {"field2": value2}]})` |
| Greater Than                                     | Retrieve documents where a specified field is greater than a certain value.                                                                                                                        | `collection.find({"field": {"$gt": value}})`            |
| Less Than                                        | Retrieve documents where a specified field is less than a certain value.                                                                                                                           | `collection.find({"field": {"$lt": value}})`            |
| Greater Than or Equal To                         | Retrieve documents where a specified field is greater than or equal to a certain value.                                                                                                            | `collection.find({"field": {"$gte": value}})`           |
| Less Than or Equal To                            | Retrieve documents where a specified field is less than or equal to a certain value.                                                                                                               | `collection.find({"field": {"$lte": value}})`           |
| Pattern Matching (Regex)                         | Retrieve documents where a specified field matches a regular expression pattern.                                                                                                                    | `collection.find({"field": {"$regex": "pattern"}})`    |
| Existence Check                                  | Retrieve documents where a specified field exists or does not exist.                                                                                                                                 | `collection.find({"field": {"$exists": True/False}})`  |
| Sorting                                          | Sort the result set based on one or more fields.                                                                                                                                                   | `collection.find().sort("field", pymongo.ASCENDING)`     |
| Limit and Skip                                    | Limit the number of documents returned and skip a specified number of documents.                                                                                                                   | `collection.find().limit(5).skip(10)`                   |
| Count                                            | Count the number of documents that match a query.                                                                                                                                                 | `collection.count_documents({"field": value})`          |
